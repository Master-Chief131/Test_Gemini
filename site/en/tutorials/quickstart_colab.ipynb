This tutorial shows you how to get started with the Gemini API using the Python SDK.

## Prerequisites

You can run this tutorial in Google Colab, which doesn't require additional environment configuration.

Alternatively, to complete this quickstart locally, see the Python guidance in [Get started with the Gemini API](https://ai.google.dev/tutorials/quickstart).

## Install the SDK

The Python SDK for the Gemini API is contained in the [`google-generativeai`](https://pypi.org/project/google-generativeai/) package. Install the dependency using pip:

```
!pip install -q -U google-generativeai
```

## Set up your API key

To use the Gemini API, you'll need an API key. If you don't already have one, create a key in Google AI Studio.

<a class="button" href="https://aistudio.google.com/app/apikey" target="_blank" rel="noopener noreferrer">Get an API key</a>

In Colab, add the key to the secrets manager under the "🔑" in the left panel. Give it the name `GOOGLE_API_KEY`. Then pass the key to the SDK:

```
# Import the Python SDK
import google.generativeai as genai
# Used to securely store your API key
from google.colab import userdata

GOOGLE_API_KEY=userdata.get('GOOGLE_API_KEY')
genai.configure(api_key=GOOGLE_API_KEY)
```

## Initialize the Generative Model

Before you can make any API calls, you need to initialize the Generative Model.

```
model = genai.GenerativeModel('gemini-pro')
```

## Generate text

```
response = model.generate_content("Write a story about a magic backpack.")
print(response.text)
```

# Scrap

```
from bs4 import BeautifulSoup
import requests
```

```
# Versiones, solo para chequeo
import bs4
print("Version de BeautifulSoup: ", bs4.__version__)
print("Version de requests: ", requests.__version__)
```

```
# Obtener el html
URL_BASE = "https://www.super99.com/despensa"
pedido_obtenido = requests.get(URL_BASE)
pedido_obtenido.status_code
html_obtenido = pedido_obtenido.text

# Parsear el html
soup = BeautifulSoup(html_obtenido, "html.parser")
```

```
type(soup)
```

```
# Extraer solo los div de los productos
divs = soup.find_all("div", class_="product-item-info")

for div in divs:
    print(div)
    print(" ")
```

```
type(divs)
```

```
# Descargar las imagenes
src_productos = soup.find_all("img", class_="product-image-photo")
for src in src_productos:
    print(src)
```

```
# Descargar las imagenes
nombre_productos = []
url_imagenes = []
for i, imagen in enumerate(src_productos):
    # Validar salida
    # print(imagen["src"])
    url_imagenes.append(imagen["src"])
    nombre_productos.append(imagen["alt"].strip())
    # r = requests.get(f"https://www.super99.com/despensa/{imagen['src']}")
    # with open(f"imagen_{i}.png", "wb") as f:
        # f.write(r.content)
```

```
# Las url tienen atributos incrustados
for i  in url_imagenes:
    print(i)
```

```
import re

# Limpiar las url
cleaned_urls = [re.sub(r'\?.*', '', url) for url in url_imagenes]

# mostrar las url limpias
for url in cleaned_urls:
    print(url)
```

```
# Limpiar los nombres de los productos
cleaned_names = [re.sub(r"/", "-", re.sub(r" ", "_", name)) for name in nombre_productos]

# mostrar los nombres limpios
for name in cleaned_names:
    print(name)
```

```
# Descargamos las imagenes y las enumeramos
for i, imagen in enumerate(cleaned_urls):
  nombre_p = cleaned_names[i]
  r = requests.get(imagen)
  with open(f"{nombre_p}.png", "wb") as f:
    f.write(r.content)
```

```
# extraer el nombre y el precio de los productos
producto_info = soup.find_all("div", class_="details-info-product")

producto_nombre = []
producto_precio = []

for info in producto_info:
    nombre = info.find("a", class_="product-item-link").text.strip()
    precio = info.find("span", class_="price-wrapper").text.strip().replace("$", "")
    producto_nombre.append(nombre)
    producto_precio.append(precio)
```

```
print("Producto/Precio")
size = len(producto_nombre)
for i in range(size):
  print(f"{producto_nombre[i]} = {producto_precio[i]}")
```
